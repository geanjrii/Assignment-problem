# Problema da Atribuio

## Descrio do problema

Considere o problema de alocao de funcionrios a tarefas ilustrado na figura abaixo. A primeira camada representa os funcionrios (ns F1, F2, F3), enquanto que a segunda camada representa tarefas a serem realizadas (ns T1, T2, T3). Cada conexo entre os ns (aresta) representa uma possibilidade de atribuio de uma tarefa a um funcionrio, tendo cada atribuio um valor diferente de aptido, como ilustrado na figura.  necessrio que __cada tarefa seja atribuda a apenas um funcionrio__ e que __cada funcionrio seja designado a apenas uma tarefa__, de tal forma que a soma das aptides utilizadas seja maximizada.

Ou seja, a soluo acima faz com que o funcionrio 1 seja alocado para a tarefa 1; funcionrio 2, para a tarefa 2; e funcionrio 3, para a tarefa 3. Portanto, o valor total das aptides utilizadas  igual a $5+7+5 = 17$

Podemos ter tambm a soluo: 



Ou seja, a soluo acima faz com que o funcionrio 1 seja alocado para a tarefa 3; funcionrio 2, para a tarefa 2; e funcionrio 3, para a tarefa 1. Portanto, o valor total das aptides utilizadas  igual a $3+7+3 = 13$, pior que o da soluo anterior.

## Modelagem do problema

#### Como entrada temos:
* $F$: conjunto de funcionrios
* $T$: conjunto de tarefas a serem realizadas pelos funcionrios (sendo |F| = |T|)
* $a_{ij} \in \mathbb{R} $: aptido do funcionrio $i \in F$ para realizar a atividade j $\in$ T

#### Variveis de deciso:
* $x_{ij} \in \{0, 1\}$ : Se o funcionrio $i \in F$ vai realizar a tarefa $j \in T$

#### Restries
* Temos que garantir que cada funcionrio seja atribudo a apenas uma tarefa
* Temos que garantir que cada tarefa seja atribuda a apenas um funcionrio

#### Funo objetivo
* Maximizar a soma das aptides utilizadas

#### A modelagem para a instncia apresentada na definio do problema seria:
$$\max 5 x_{1,1} + 4 x_{1, 2} + 3 x_{1, 3} + 8 x_{2, 1} + 7 x_{2, 2} + 6 x_{2, 3} + 3 x_{3, 1} + 7 x_{3, 2} + 5 x_{3, 3}$$
$S.a.:$
$$x_{1, 1} + x_{1, 2} + x_{1, 3} = 1 $$
$$x_{2, 1} + x_{2, 2} + x_{2, 3} = 1 $$
$$x_{3, 1} + x_{3, 2} + x_{3, 3} = 1 $$
$$x_{1, 1} + x_{2, 1} + x_{3, 1} = 1 $$
$$x_{1, 2} + x_{2, 2} + x_{3, 2} = 1 $$
$$x_{1, 3} + x_{2, 3} + x_{3, 3} = 1 $$
$$ x_{1,1}, x_{1, 2}, x_{1, 3}, x_{2, 1}, x_{2, 2}, x_{2, 3}, x_{3, 1}, x_{3, 2}, x_{3, 3} \in \{0, 1\}$$

## Soluo para o problema utilizando Python MIP

```python
from mip import *

# Matriz de Aptides
fit_3 = [[5, 4, 3],  # F1-T1, F1-T2, F1-T3
         [8, 7, 6],  # F2-T1, F2-T2, F2-T3
         [3, 7, 5]]  # F3-T1, F3-T2, F3-T3

# Nmero de funcionrios e de tarefas
n_emp_3 = n_task_3 = 3

# Instanciando o modelo e atribuindo que a funo objetivo  de maximizao
model_3 = Model(sense=MAXIMIZE, solver_name=CBC)

# Fazendo list comprehension para criar matriz de variaveis com a mesma 
# estrutura do fit acima e iterando 3x nos dois for para representar 3 func. e 3 tarefas
x = [[model_3.add_var(var_type="BINARY", name = f"x_{i+1}_{j+1}") for j in range(n_task_3)]
            for i in range(n_emp_3)]

# Funo objetivo
model_3.objective = fit_3[0][0]*x[0][0] + fit_3[0][1]*x[0][1] + fit_3[0][2]*x[0][2] + \
                    fit_3[1][0]*x[1][0] + fit_3[1][1]*x[1][1] + fit_3[1][2]*x[1][2] + \
                    fit_3[2][0]*x[2][0] + fit_3[2][1]*x[2][1] + fit_3[2][2]*x[2][2]

# Restries

# Funcionrios
model_3 += x[0][0] + x[0][1] + x[0][2] == 1
model_3 += x[1][0] + x[1][1] + x[1][2] == 1
model_3 += x[2][0] + x[2][1] + x[2][2] == 1

# Tarefas
model_3 += x[0][0] + x[1][0] + x[2][0] == 1
model_3 += x[0][1] + x[1][1] + x[2][1] == 1
model_3 += x[0][2] + x[1][2] + x[2][2] == 1 

status = model_3.optimize()

print("Status = ", status)
print("Solution value = ", model_3.objective_value)

print("Solution:")
for v in model_3.vars: 
    if v.x > 0.00001: 
        print(v.name, " = ", v.x)
```

#### Assim, temos uma soluo tima com valor 18:

## Assim, a modelagem geral para o problema da atribuio seria:

* As restries dos funcionrios garantem que apenas uma tarefa seja designada para cada funcionrio.

* As restries das tarefas garantem que apenas um funcionrio seja designado para cada tarefa.

* Essas duas restries torna a relao 1 para 1.

* Considere $n$ o nmero de funcionrios (que tambm  igual ao nmero de tarefas)

$$\max a_{1,1}x_{1,1} + a_{1,2}x_{1,2}+...+a_{1,n}x_{1,n}+a_{2,1}x_{2,1}+a_{2,2}x_{2,2}+...+a_{2,n}x_{2,n} + ... + a_{n,1}x_{n,1}+a_{n,2}x_{n,2}+...+a_{n,n}x_{n,n}$$
$S.a.:$
$$\space$$
$$Funcionrios:$$
$$x_{1, 1} + x_{1, 2} + x_{1, 3} +...+ x_{1, n}= 1 $$
$$x_{2, 1} + x_{2, 2} + x_{2, 3} +...+ x_{2, n}= 1 $$
$$\vdots$$
$$x_{n, 1} + x_{n, 2} + x_{n, 3} +...+ x_{n, n}= 1 $$
$$\space$$
$$Tarefas:$$
$$x_{1, 1} + x_{2, 1} + x_{3, 1} +...+x_{n, 1}= 1 $$
$$x_{1, 2} + x_{2, 2} + x_{3, 2} +...+x_{n, 2}= 1 $$
$$\vdots$$
$$x_{1, n} + x_{2, n} + x_{3, n} +...+ x_{n, n}= 1 $$
$$ x_{1,1}, x_{1,2},...,x_{1,n},x_{2,1},x_{2,2},...,x_{2,n},..., x_{n,1},x_{n,2},...,x_{n,n} \in \{0, 1\}$$

###Escrita compacta do modelo geral:

1.   Item da lista
2.   Item da lista


$$\max \sum_{i \in F} \sum_{j \in T} a_{i,j}x_{i,j}$$
$$S.a.:$$
$$ \sum_{i \in F} x_{i, j} = 1, \space \forall j \in T$$
$$ \sum_{j \in T} x_{i, j} = 1, \space \forall i \in F$$
$$x_{i, j} \in \{0, 1\}, \space\space \forall i \in F, \forall j \in T$$

### Logo, temos uma funo geral para o problema de atribuio:

```python
def create_model(n_emp: int, n_task: int, fit: List[List[int]]):
    """
    Funo que criar o modelo com todas as restries e funo objetivo.

    Parameters:
    -----------
    n_emp: int
        Nmero de empregados/funcionrios
    n_task: int
        Nmero de tarefas
    fit: List[List[int]]
        Matriz de tamanho n_emp x n_task contendo a aptido de cada 
        funcionrio para cada tarefa.

    Returns:
    --------
    model: mip.Model
        Modelo j instanciado e pronto para ser otimizado
    """
    
    # Instanciando o modelo e atribuindo que a funo objetivo  para maximizar
    model = Model(sense=MAXIMIZE, solver_name=CBC)
    
    # Fazendo list comprehension para criar matriz de variaveis
    x = [
        [model.add_var(var_type="BINARY", name = f"x_{i+1}_{j+1}") for j in range(n_task)]
            for i in range(n_emp)]

    # Funo objetivo
    model.objective = xsum(fit[i][j]*x[i][j]
                           for i in range(n_emp) for j in range(n_task))
    
    # Restries

    # Funcionrios
    for i in range(n_emp):
        model += xsum(x[i][j] for j in range(n_task)) == 1

    # Tarefas
    for j in range(n_task):
        model += xsum(x[i][j] for i in range(n_emp)) == 1 
    
    return model
```

```python
# Utilizando a funo acima para o modelo com 3 funcionarios e 3 tarefas
model3 = create_model(3, 3, fit_3)

status = model3.optimize()

print("Status = ", status)
print("Solution value = ", model3.objective_value)

print("Solution:")
for v in model3.vars: 
    if v.x > 0.00001: 
        print(v.name, " = ", v.x)
```
